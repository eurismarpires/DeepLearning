```python
from keras.layers import Conv2D, MaxPooling2D, Activation
from keras.models import Sequential
from keras.initializers import RandomUniform
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import cv2  
```

```python
cat = cv2.imread('cat.jpg')
plt.imshow(cat)
```

```python
cat.shape
```

```python
model = Sequential()
#inicializer = RandomUniform(minval=0.0, maxval=255.0, seed=None)
#model.add(Conv2D(3,(3,3),input_shape=cat.shape, kernel_initializer=inicializer))
model.add(Conv2D(3,(3,3),input_shape=cat.shape,activation='relu'))
```

```python
cat_batch = np.expand_dims(cat,axis=0)
conv_cat = model.predict(cat_batch)
```

```python
conv_cat.shape
```

```python
conv_cat = np.squeeze(conv_cat, axis=0)
```

```python
plt.imshow(conv_cat)
```

```python
def visualize_cat(model, cat):
    cat_batch = np.expand_dims(cat,axis=0)
    conv_cat = model.predict(cat_batch)
    conv_cat = np.squeeze(conv_cat, axis=0)
    print(conv_cat.shape)
    plt.imshow(conv_cat)
    
visualize_cat(model, cat)
```

```python
model = Sequential()
model.add(Conv2D(3, (10,10),input_shape=cat.shape,activation='relu'))
```

```python
visualize_cat(model, cat)
```

```python
def imprime_gatinho_fofo(model, cat):
    cat_batch = np.expand_dims(cat,axis=0)
    conv_cat2 = model.predict(cat_batch)
    conv_cat2 = np.squeeze(conv_cat2, axis=0)
    print(conv_cat2.shape)
    conv_cat2 = conv_cat2.reshape(conv_cat2.shape[:2])
    print(conv_cat2.shape)
    plt.imshow(conv_cat2)
    
model = Sequential()
model.add(Conv2D(1,(3,3),input_shape=cat.shape))
```

```python
imprime_gatinho_fofo(model, cat)
```

```python
model = Sequential()
model.add(Conv2D(1,(15,15),input_shape=cat.shape))

imprime_gatinho_fofo(model, cat)
```

```python
model = Sequential()
model.add(Conv2D(1,(3,3),input_shape=cat.shape))
model.add(Activation('relu'))

imprime_gatinho_fofo(model, cat)
```

```python
model = Sequential()
model.add(Conv2D(3,(3,3),input_shape=cat.shape))
model.add(Activation('relu'))

visualize_cat(model, cat)
```

```python
model = Sequential()
model.add(Conv2D(1,(3,3),input_shape=cat.shape))
model.add(MaxPooling2D(pool_size=(5,5)))

imprime_gatinho_fofo(model, cat)
```

```python
model = Sequential()
model.add(Conv2D(1,(3,3),input_shape=cat.shape))
model.add(MaxPooling2D(pool_size=(5,5)))
model.add(Conv2D(1,(3,3),input_shape=cat.shape))
model.add(MaxPooling2D(pool_size=(5,5)))
imprime_gatinho_fofo(model, cat)
```

```python
model = Sequential()
model.add(Conv2D(1,(3,3),input_shape=cat.shape))
model.add(MaxPooling2D(pool_size=(5,5)))
imprime_gatinho_fofo(model, cat)
```

```python
model = Sequential()
model.add(Conv2D(1,(3,3),input_shape=cat.shape))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(5,5)))
imprime_gatinho_fofo(model, cat)
```

```python
model = Sequential()
model.add(Conv2D(3,(3,3),input_shape=cat.shape))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(5,5)))

visualize_cat(model, cat)
```

```python
model = Sequential()
model.add(Conv2D(1,(3,3),input_shape=cat.shape))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(3,3)))
model.add(Conv2D(1,(3,3),input_shape=cat.shape))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(3,3)))

imprime_gatinho_fofo(model, cat)
```

```python
model = Sequential()
model.add(Conv2D(3,(3,3),input_shape=cat.shape))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(3,3)))
model.add(Conv2D(1,(3,3),input_shape=cat.shape))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

imprime_gatinho_fofo(model, cat)
```

```python
model = Sequential()
model.add(Conv2D(3,(3,3),input_shape=cat.shape))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(3,3)))
model.add(Conv2D(3,(3,3),input_shape=cat.shape))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

visualize_cat(model, cat)
```
