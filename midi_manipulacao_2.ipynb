```python
f = "http://www.classicalguitarmidi.com/subivic/Brouwer_Etude_No1.mid"
midi_stream = converter.parse(f)
```

```python
Image(filename=midi_stream.write('lily.png'))
```

```python
lista_notas = []
for m in midi_stream.measures(1,None):
    time_sig = m.timeSignature           
    for nr in m.flat.notesAndRests:
        offset_loc = nr.offset               
        notas = []               
        if isinstance(nr, note.Note):
            pitch = nr.pitch.midi            
            notas.append(pitch)
        else:
            notas = [pitch.midi for pitch in nr]        
        linha =(nr.offset,notas)                      
        nt = (nr.offset,notas)
        #nt = (notas,)
        #print(type(nt))
        lista_notas.append(nt)
tempo_anterior = 0.0
l_notas = []
for i, a in enumerate(lista_notas):    
    tempo = lista_notas[i+1][0] - lista_notas[i][0] if i+1 < len(lista_notas) else lista_notas[i][0] - lista_notas[i-1][0]        
    t = (lista_notas[i]) + (tempo,)    
    l_notas.append(t)        
    #print(i,a)
```
